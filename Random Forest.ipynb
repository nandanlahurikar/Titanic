```python
from sklearn.ensemble import RandomForestRegressor 
from sklearn.metrics import roc_auc_score
import pandas as pd
```

```python
X = pd.read_csv('train.csv')
Y = X.pop("Survived")
```

```python

X.describe()
```

```python
X.head()
```

```python
#fill missing values from age catagory

X["Age"].fillna(X.Age.mean(), inplace=True)
X.describe()
```

```python
# just get numerical values by selecting only the varibales tharare not 'object' data set

numeric_variable = list(X.dtypes[X.dtypes != "object"].index) # [X.dtypes != "object"]
print (numeric_variable)
X[numeric_variable].head()
```

```python
model = RandomForestRegressor(n_estimators=100, oob_score=True,random_state=42)
model.fit(X[numeric_variable],Y)
```

```python
model.oob_score_
#model.
```

```python
Y_oob = model.oob_prediction_
print("c-stat", roc_auc_score(Y,Y_oob))

```

```python
# make analysis on catagoricle values
def describ_cat(X):
    # return results for catgorikal values only
    from IPython.display import display, HTML
    display(HTML(X[X.columns[X.dtypes=="object"]].describe().to_html()))
```

```python
describ_cat(X)
```

```python
#drop varibles that dont feel relevant
X.drop(["Name","Ticket","PassengerId"], axis=1, inplace=True)
```

```python
# change cabin variblea are 
# program changes each cabin value with initial letter and return none for missing values
def clean_cabin(x):
    try:
        return x[0]
    except TypeError:
        return "None"
X["Cabin"] = X.Cabin.apply(clean_cabin)
```

```python
X.Cabin
```

```python
Catagorical_variables = ['Sex','Cabin','Embarked']

for variable in Catagorical_variables:
    #fill missing data with "missing"
    X[variable].fillna("Missing", inplace=True)
    #create array of dummies
    dummies = pd.get_dummies(X[variable], prefix=variable)
    X = pd.concat([X,dummies],axis=1)
    X.drop([variable],axis=1,inplace=1)
    
```

```python
X
```

```python
model = RandomForestRegressor(n_estimators=100, oob_score=True,random_state=42)
model.fit(X,Y)
```

```python

print("c-stat", roc_auc_score(Y,model.oob_prediction))

```
